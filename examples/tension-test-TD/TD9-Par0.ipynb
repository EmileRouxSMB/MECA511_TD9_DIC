````python
# If missing gdown, install it using pip
# pip install gdown
import gdown

# URL of the folder
folder_url = "https://drive.google.com/drive/folders/1MT_qLXPw94TToyeKtjPDahX1sGih4I63?usp=sharing"

# Dossier de sortie = dossier courant
output_dir = os.getcwd()   # équivaut à "."

# Téléchargement de tous les fichiers du dossier Google Drive
gdown.download_folder(
    url=folder_url,
    output=output_dir,
    quiet=False,
    use_cookies=False
)

print("Téléchargement terminé dans :", output_dir)
````
